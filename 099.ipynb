# Reverse name and change to upper case


````python
def reverse_and_uppercase_name(name):
    # Split the name into first and last name
    parts = name.split()

    # Check if there are exactly two parts (first name and last name)
    if len(parts) == 2:
        first_name, last_name = parts
        # Convert both to uppercase and reverse the order
        return f"{last_name.upper()} {first_name.upper()}"
    else:
        return "Invalid input. Please provide a first and last name."

# Example usage
name = "Stephen Hawking"
result = reverse_and_uppercase_name(name)
print(result)
````

# Convert to all upper case

````python
def convert_to_uppercase(name):
    return name.upper()

# Example usage
name = "Myrlie Evers-Williams"
uppercase_name = convert_to_uppercase(name)
print(uppercase_name)

````

# 245 Formatting

````python
def format_book_entry(text):
    # Split the title and author sections by the slash '/'
    if '/' not in text:
        return "Invalid input. Please provide text in the format: 'Title:Subtitle/Author'"

    title_section, author_section = text.split('/')

    # Split title_section by colon ':' to get the title and subtitle
    if ':' not in title_section:
        return "Invalid input. Title and subtitle must be separated by a colon."

    title, subtitle = title_section.split(':', 1)

    # Assume author_section can contain both primary and real name in parentheses
    author_section = author_section.strip()

    # Split the author_section into author name and additional info in parentheses if it exists
    if '(' in author_section and ')' in author_section:
        author, additional_info = author_section.split('(', 1)
        additional_info = additional_info.strip(' )')  # Remove surrounding spaces and parentheses
        author = author.strip()
        formatted_author = f"{author} ({additional_info})"
    else:
        formatted_author = author_section.strip()

    # Construct the final formatted string
    formatted_entry = f"{title.strip()} : ǂb {subtitle.strip()} / ǂc {formatted_author}"

    return formatted_entry

# Example usage
input_text = "The heart of the world:The isles of the gods volume 2/ Amie Kaufman"
formatted_output = format_book_entry(input_text)
print(formatted_output)
````
